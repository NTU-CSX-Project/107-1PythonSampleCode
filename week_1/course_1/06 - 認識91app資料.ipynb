### 使用 Pandas 的 `read_csv()` 方法讀取 91APP 提供的資料 ( csv檔 )

1. 此時你必須逐一檢查並清理這些資料，使其符合合理的資料型態。(一開始讀入的格式不見得合理，因為它只是原封不動的讀入資料)
2. 這邊舉第二個變數 DateId 為例，原先的資料在這邊被誤解成數值 (int)，但我們希望它是日期格式 (Date)，因此我們會使用 `to_datetime()` 方法進行修正。( 這邊只需要先知道分析資料前，我們總是會依照自己腦中的邏輯去把資料進行整理 )

### 使用 `to_datetime()` 把 DateId 轉成日期格式並存成新的變數 new_DateId 的一個例子。

```python
df.info()
```

```python
df.head()
```

## △ 2. 資料選取 (Slice) 與過濾 (Filter)


### △ 2.1 資料選取的方式分成 3 種 :
1. 選取整個 column 
2. 選取整個 row
3. 一起選取

### 使用中括號 [ ] 選取 column 名稱 :
下面的例子就是選取 "購買數量 (Quantity)" 的欄位，並且只印出前 5 筆。

```python
df['Quantity'].head()
```

### 使用 `loc()` 方法透過 index 選取整行 row :
下面的例子是透過 index = 3 去選取第 4 筆資料的位置 ( index 從 0 開始 )。<br>
**( 注意 : `loc()` 方法內的 index 不會減 1。)**

```python
df.loc[3]
```

### `使用 loc()` 方法同時選取 column 和 row :
下面的例子是選取 "購買數量 (Quantity)" 欄位的 index = 1000 到 1005 筆資料。<br>
**( 注意 : `loc()` 方法內的 index 不會減 1。)**

```python
df.loc[1000:1005, 'Quantity']
```

### 也可以同時選取多個 column :
只要一起放在 python 的任意可跌代對象就可以了，像是 list, tuple 等等。

```python
df.loc[1000:1005, ['new_DateId', 'Quantity']]
```

### △ 2.2 資料過濾 (Filter) :
### 中括號 [ ] 除了能幫我們選擇 column 外，還能幫助我們進行資料的過濾塞選。
下面的例子是選取 "購買數量 (Quantity)" 超過 1 件商品的所有資料。

```python
df[df['Quantity'] > 1]
```

## △ 3. 資料分組 (groupby)
### groupby() 功能可以把值相同的資料合併再一起，並且進行你給他的後續運算工作。( 直接用下面舉例比較清楚 )
完整運作原理為下圖，把相同值的 key 後面的 data 併在一起，並進行你給他的運算動作，此圖給的動作為 `sum()` 因此它就是把各 key 的資料分別加在一起。

<img src='https://i.stack.imgur.com/sgCn1.jpg'>
我們以我們的例子 (下方) 再一次理解 `groupby()` 的運作方式，這邊的 key 就是對應我們在 new_DateId 中的各個日期 (ex. 2017-01-01, 2017-01-02) <br>
而 data 其實就是對應我們的 Quantity 欄位，在 split 後最後又再被我們 sum() 在一起，意思就是我們簡單整理了一個 **每日的商品銷售量** 
表格，結果如下 :

```python
data = df.groupby(by='new_DateId')['Quantity'].sum()
data.head(10)
```

### 最後我們直接拿剛剛整理好的表格用 pandas 內建的簡易畫圖工具看看結果
( p.s. 為了在 jupyter notebook 中看到圖片，必需要加的一行 `%matplotlib inline`，不然只會返回 object。 )

```python
%matplotlib inline
data.plot()  # 會默認第一個變數是 X 軸，第二個變數是 Y 軸。
```

### 這樣可以清楚看到 2017 1月到 2018 1月銷售數量的狀況了~
